```python
#Getting state name from St 
data_2010['state_name'] = data_2010['St'].map(state_dict_2)
#Removing records where latest remmoval date has NA
data_2010= data_2010.dropna(subset=['LatRemDt'])
#only keeping those records which were removed in that particular years 
data_2010['LatRemDt'] = pd.to_datetime(data_2010['LatRemDt'])
data_2010['Year_removal'] = data_2010['LatRemDt'].dt.year.astype(int)
data_2010= data_2010[data_2010['Year_removal']==2010]


#getting the counts of removals

data_2010_overall  = data_2010.groupby('state_name')['RecNumbr'].nunique().reset_index(name='all_removals_state_2010')
population = population_time[['State', '2010']]
data_2010_overall = data_2010_overall.merge(population,left_on = 'state_name', right_on = 'State', how = 'right')
data_2010_overall['ror_state_2010'] = (data_2010_overall['all_removals_state_2010']/ data_2010_overall['2010']) * 1000



data_2010_housing_only  = data_2010[(data_2010['HOUSING'] == 1) 
 & (data_2010['NEGLECT'] == 0) 
 & (data_2010['PHYABUSE'] == 0) 
 & (data_2010['SEXABUSE'] == 0) 
 & (data_2010['AAPARENT'] == 0)
& (data_2010['DAPARENT'] == 0)
& (data_2010['AACHILD'] == 0)
& (data_2010['DACHILD'] == 0)
& (data_2010['CHILDIS'] == 0)
& (data_2010['CHBEHPRB'] == 0)
& (data_2010['PRTSDIED'] == 0)
& (data_2010['PRTSJAIL'] == 0)
& (data_2010['NOCOPE'] == 0)
& (data_2010['ABANDMNT'] == 0)
& (data_2010['RELINQSH'] == 0)]

data_2010_housing_only = data_2010_housing_only.groupby(['state_name'])['RecNumbr'].nunique().reset_index(name='housing_removals_state_2010')
data_2010_housing_only = data_2010_housing_only.merge(population_time,left_on = 'state_name', right_on = 'State', how = 'right')


#Merging those 

data_2010_housing_only['ror_housing_state_2010'] = (data_2010_housing_only['housing_removals_state_2010']/ data_2010_housing_only['2010']) * 1000
data_2010_housing_only = data_2010_housing_only[['State','housing_removals_state_2010','2010','ror_housing_state_2010']]

data_2010_all = data_2010_overall.merge(data_2010_housing_only, on = ['State','2010'])


data_2010_all = data_2010_all.drop(columns = 'state_name')
data_2010_all.head()
```
